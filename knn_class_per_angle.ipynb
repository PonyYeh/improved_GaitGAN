```python
#目的地是要看 cross- view
# %load knn_class_per_angle.py
%load_ext autoreload
%autoreload 2
import cv2
import os
import torch as th
from model_siGAN import NetG, NetD, NetA
from dataset2Loader import loadImage
import visdom
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

epoch = 'total_multimes'
# checkpoint_36 = './Transform_Model/Model_64x64_TripletGAN_36_trial1/lowest_snapshotModel_64x64_TripletGAN_36_trial1_598.t7'
checkpoint_90 = './Transform_Model/Model_64x64_TripletGAN_90_trial1/lowest_snapshotModel_64x64_TripletGAN_90_trial1_613.t7'
# checkpoint_144 = './Transform_Model/Model_64x64_TripletGAN_144_trial1/lowest_snapshotModel_64x64_TripletGAN_144_trial1_613.t7'

Transform_dir_36 = './Transform_64x64_allseq_36'
Transform_dir_144 = './Transform_64x64_allseq_144'

Data_Dir = '../GaitRecognition/GEI_CASIA_B/gei/'
Model_Name = 'Model_64x64_Triplet_36_90_144'
Model_dir = './Transform_Model/'+ Model_Name


netg = NetG(nc=1)
device = th.device("cuda:1")
netg = netg.to(device)
checkpoint = th.load(checkpoint_90)
netg.load_state_dict(checkpoint['netG'])
netg.eval()

result = np.zeros((12, 12))
angles_gallery = ['000', '018', '036', '054', '072',
                  '090',
                  '108', '126', '144', '162', '180']
angles_probe = angles_gallery
ix = 0
iy = 0
diagonal= []
for idx_p, p_ang in enumerate(angles_gallery):
    print('\np_ang',p_ang)

    ix = 0
    avg = []
    for g_ang in angles_probe:      
        pid = 63
        X = []
        y = []        
        #看要不要改成gallery角度是只有90度，比較符合情境，
        for cond in ['nm-01', 'nm-02', 'nm-03', 'nm-04']:
            for p in range(pid, 125):                
                #setting1
                if(g_ang in ['000','018','036']):
                    path = Transform_dir_36 + '/%03d-%s-%s.png' % ( p, cond, g_ang)
                elif(g_ang in ['144','162','180']):
                    path = Transform_dir_144 + '/%03d-%s-%s.png' % ( p, cond, g_ang)
                elif(g_ang in['054','072','090','108','126']):
                    path = Data_Dir+'/%03d/%s/%03d-%s-%s.png' % (p, cond, p, cond, g_ang)
                if not os.path.exists(path):
                    continue                   
                img = cv2.imread(path, 0)               
                img = cv2.resize(img, (64, 64))
                X.append(img)
                y.append(p-63)    
        nbrs = KNeighborsClassifier(n_neighbors=3, p=1, weights='distance')
        X = np.asarray(X).astype(np.float32) #(2704, 64, 64)
        X = th.from_numpy(2 * (X/255) - 1).unsqueeze(1) # torch.Size([2704, 1, 64, 64])
        X = X.to(device).to(th.float32)
        with th.no_grad():
            fake_X, encoder_X = netg(X)
        fake_X = fake_X.cpu().numpy() #(2704, 1, 64, 64)
        fake_X = fake_X.reshape(-1,64*64) #(2704, 12288)
        fake_X = (fake_X + 1) / 2 * 255
        nbrs.fit(fake_X, y)

        testX = []
        testy = []
        pid = 63
        for cond in ['nm-05', 'nm-06']:
            for p in range(pid, 125):
                if(p_ang in ['000','018','036']):
                    path1 = Transform_dir_36 + '/%03d-%s-%s.png' % ( p, cond, p_ang)
                elif(p_ang in ['144','162','180']):
                    path1 = Transform_dir_144 + '/%03d-%s-%s.png' % ( p, cond, p_ang)
                elif(p_ang in['054','072','090','108','126']):
                    path1 = Data_Dir+'/%03d/%s/%03d-%s-%s.png' % (p, cond, p, cond, p_ang)
#                 path1 = Data_Dir+'/%03d/%s/%03d-%s-%s.png' % (p, cond, p, cond, p_ang)
                if not os.path.exists(path1):
                    continue
                img = cv2.imread(path1, 0)
                img = cv2.resize(img, (64, 64))
                testX.append(img)
                testy.append(p-63)
            
        testX = np.asarray(testX).astype(np.float32) 
        testX = th.from_numpy(2 * (testX/255) - 1).unsqueeze(1) # torch.Size([2704, 1, 64, 64])
        testX = testX.to(device).to(th.float32)
        with th.no_grad():
            fake_testX,encoder_testX= netg(testX)
        fake_testX = fake_testX.cpu().numpy()
        fake_testX = fake_testX.reshape(-1,64*64)
        fake_testX = (fake_testX + 1) / 2 * 255
        s = nbrs.score(fake_testX, testy)
        result[iy][ix] = s
        
        print('g_ang',g_ang,s)
        if(p_ang == g_ang):
            diagonal.append(result[idx_p][idx_p])  
        ix += 1

    for c in range(11):
        if(idx_p!=c):
            avg.append(result[iy][c])
    result[iy][ix] = np.mean(avg)      
    iy += 1
result[11][11]= np.mean(diagonal)
print(result)

np.savetxt(Model_dir+"/PerAngle_embedding_knn_ProbeNM_"+str(epoch)+".csv", result, delimiter="," ,fmt='%1.4f')
```
